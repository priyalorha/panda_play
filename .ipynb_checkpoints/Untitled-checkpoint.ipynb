```python
import pandas as pd
```

```python
df = pd.read_csv('banking.csv')
```

```python
def fix_common_errors(s):
    if 'ACCl' in s:
        _,x =s.split('ACCl')
        y = ''
        common_errors = {'l': '1', 'z': '2', 's': '5', 'o':'0', 'g':'6'}
        if str(x).isnumeric():
            return s
        else:
            for i in range(len(x)):
                y += common_errors.get(x[i], x[i])
       
        return "ACCl" + y
    return s
```

```python
def remove_dollar_symbol(s):
    return float(s.replace('$', ''))
```

#### DATA CLEANING

#### FIX ACCOUNT NUMBER

```python
df['Account Number'] = df['Account Number'].apply(fix_common_errors)
```

#### REMOVE DOLLAR SYMBOL

```python
df['Amount'] = df['Amount'].apply(remove_dollar_symbol)
```

```python
df['credit'] = 0
df['debit']=0
df['remarks']=None
```

```python
#spiliting amount in credit and debit
```

```python
def seg_data(df):
    
    if df['Transaction Type'] in ['ATM Withdrawal', 'Withdrawal','Direct Debit', 'Card Payment']:
        df['debit'] = df['Amount'] if df['Amount'] > 0  else -1 * df['Amount']
    elif df['Transaction Type'] in ['Deposit']:
        df['credit'] = df['Amount'] if df['Amount'] > 0  else -1 * df['Amount']
    elif df['Transaction Type'] == 'Online Transfer':
        if df['Description'] == 'Salary':
            df['credit'] = df['Amount'] if df['Amount'] > 0  else -1 * df['Amount']

        elif df['Amount']>0:
            df['credit'] = df['Amount']
        else:
            df['debit'] = df['Amount']*-1
        
        
    
        
    
    return df
```

```python
df=df.apply(lambda x : seg_data(x), axis =1)
```

```python
grouped = df.groupby(['Account Number', 'Transaction Type','Description'])
grouped_avg = grouped['Amount'].transform('mean')
df['remark'] = df.apply(lambda row: 'Seem odd to me' if row['Amount'] > grouped_avg[row.name] or row['Amount'] < grouped_avg[row.name] else '', axis=1)
```
